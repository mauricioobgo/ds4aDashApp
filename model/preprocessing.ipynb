```python
import pandas as pd
```

```python
EVA_06_to_18=pd.read_csv('./data/EVA_2001_2018.csv')
EVA_19_to_20=pd.read_csv('./data/EVA_2019_2020.csv')

# Renombrar columnas de la primer base de datos para poderlos unir ya que tienen diferentes nombres de columna

EVA_06_to_18=EVA_06_to_18.rename(columns={'CÓD. \nDEP.':'Codigo del departamento', 'DEPARTAMENTO':'Departamento', 'CÓD. MUN.':'Código del Municipio',
                              'MUNICIPIO':'Municipio','GRUPO \nDE CULTIVO':'Grupo cultivo según especie', 'SUBGRUPO \nDE CULTIVO':'Subgrupo cultivo según especie',
                              'CULTIVO':'Cultivo', 'AÑO':'Año', 'PERIODO':'Periodo','Área Sembrada\n(ha)':'Area Sembrada (ha)', 'Área Cosechada\n(ha)':'Area Cosechada (ha)',
                              'Producción\n(t)':'Producción (t)','Rendimiento\n(t/ha)':'Rendimiento (t/ha)', 'ESTADO FISICO PRODUCCION':'Estado físico de la producción',
                              'NOMBRE \nCIENTIFICO':'Nombre científico', 'CICLO DE CULTIVO':'Ciclo del cultivo'}).drop('DESAGREGACIÓN REGIONAL Y/O SISTEMA PRODUCTIVO',axis=1)

# Unir las bases de datos
eva=pd.concat([EVA_06_to_18, EVA_19_to_20],ignore_index = True)
```

```python
eva = eva.rename(columns={'Codigo del departamento':'dept_code', 'Departamento':'dept', 'Código del Municipio':'mun_code', 'Municipio':'mun', 'Grupo cultivo según especie':'group', 'Subgrupo cultivo según especie':'subgroup',
 'Subgrupo cultivo según especie':'subgroup', 'Cultivo':'crop', 'Año':'year', 'Periodo':'period', 'Area Sembrada (ha)':'area_seeded', 'Area Cosechada (ha)':'area_harvested',
 'Producción (t)':'production', 'Rendimiento (t/ha)':'yield', 'Estado físico de la producción':'production_status', 'Nombre científico':'scientific_name', 'Ciclo del cultivo':'cycle'})
```

```python
# Estandarización de las columnas con texto del dataframe

eva['dept']=eva['dept'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
eva['mun']=eva['mun'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
eva['group']=eva['group'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
eva['subgroup']=eva['subgroup'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
eva['crop']=eva['crop'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
eva['production_status']=eva['production_status'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
eva['scientific_name']=eva['scientific_name'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
eva['cycle']=eva['cycle'].str.lower().str.strip().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

for code_i in eva['mun_code'].unique():
    indexes = eva[eva['mun_code'] == code_i].index
    muns = eva[eva['mun_code']==code_i]['mun'].unique()
    muns = [str(x) for x in muns if str(x) != 'nan']
    selected_mun = min(muns, key=len)
    eva.loc[indexes, 'mun'] = selected_mun


eva.loc[eva[eva['area_harvested']==0].index, 'yield'] = 0

eva.loc[eva[eva['area_harvested']==0].index, 'production'] = 0

eva2=eva

eva2['yield']=eva2['production']/eva2['area_harvested']

eva2

eva[(eva['crop']=='aguacate')&(eva['mun_code']==54051)].sort_values(by='period',ascending=False).head(20)
```
